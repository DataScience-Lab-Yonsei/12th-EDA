```python
import pandas as pd
import numpy as np
more=pd.read_excel(r'/content/박스오피스 연도별 자료 필터링.xlsx',sheet_name=8) #2016년
df=pd.read_excel(r'/content/combined_2024.xlsx',sheet_name=0) #2024년

df['date'] = pd.to_datetime(df['date'].str.replace('.', ''), format='%Y%m%d', errors='coerce')
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# 영화별 개봉일 전후 1주일 데이터 필터링
less = df[df.apply(lambda row: row['release_date'] - pd.Timedelta(days=7) <= row['date'] <= row['release_date'] + pd.Timedelta(days=7), axis=1)]

del less['comment']
del less['date']
del less['type']
del less['release_date']

grouped_less = less.groupby('keyword').agg({
    'rating': 'mean',
    'up': 'mean',
    'down': 'mean'
}).reset_index()

# 평균값만 포함된 데이터프레임 생성
final_grouped_less = grouped_less.rename(columns={
    'rating': 'rating_mean',
    'up': 'up_mean',
    'down': 'down_mean'
})


final_grouped_less.columns=["영화명","rating_mean","up_mean","down_mean"]
print(final_grouped_less)

# Merge를 이용하여 합치기 (outer join)
merged_df = pd.merge(more, final_grouped_less, on='영화명', how='outer')

print("Merged DataFrame (outer join):")
print(merged_df)

merged_df_replaced = merged_df.replace('', np.nan)##
merged_df_filled = merged_df_replaced.fillna('NA')
merged_df_filled.to_excel('merged_2024.xlsx',index=False) ##
```
