```python
#3. running_time는 평균아니면 중간값 분포보고 결정, 중간값으로 결정
import matplotlib.pyplot as plt
import seaborn as sns
df_cleaned_2 = df_cleaned.dropna(subset=['running_time']) #running_time의 결측치 제거

df_cleaned_2['running_time'] = df_cleaned_2['running_time'].astype(str).str.replace('분', '', regex=False).astype(float)

plt.figure(figsize=(12, 6))
sns.histplot(df_cleaned_2['running_time'], bins=20, kde=True)
plt.title('Running Time Distribution')
plt.xlabel('Running Time (minutes)')
plt.ylabel('Frequency')
plt.show()

# 시각화: 박스 플롯
plt.figure(figsize=(12, 6))
sns.boxplot(x=df_cleaned_2['running_time'])
plt.title('Running Time Box Plot')
plt.xlabel('Running Time (minutes)')
plt.show()

print(df_cleaned_2['running_time'].mean())
print(df_cleaned_2['running_time'].median())
```

```python
from scipy import stats
running_time = df_cleaned_2['running_time']

# Shapiro-Wilk 검정
shapiro_test = stats.shapiro(running_time)

print('Shapiro-Wilk 검정 결과')
print('검정 통계량:', shapiro_test.statistic)
print('p-값:', shapiro_test.pvalue)

if shapiro_test.pvalue>0.05:
  print("정규분포를 따른다.")
else:
  print("running_time은 정규분포를 따르지 않는다.")

#정규성 따르지않으므로 중간값으로 결측치 채우기=105분

df_cleaned.loc[:, 'running_time'] = df_cleaned['running_time'].fillna("105분")
#missing = df_cleaned['running_time'].isnull().sum() #결측치 개수 확인

df_cleaned['running_time'].isnull().sum()
```

```python
#여기 추가

#df_cleaned['대표국적'] = df_cleaned['대표국적'].replace({'일본': '한국': 1, '미국': 2, '스페인': 2})

def convert_country(country):
    if country == '한국':
        return '한국'
    elif country in ['일본', '중국', '인도', '홍콩', '대만', '태국', '필리핀']:
        return '한국 외 아시아'
    elif country == '미국':
        return '미국'
    elif country in ['영국', '프랑스', '독일', '캐나다', '호주', '이탈리아', '스페인', '네덜란드']:
        return '미국 외 서구 국가'
    else:
        return '그 외'

# 값 변경
df_cleaned['대표국적'] = df_cleaned['대표국적'].apply(convert_country)

print(df_cleaned)

```

```python
#이거도 추가
def convert_genre(genre):
    if genre == '애니메이션':
        return '애니메이션'
    elif genre in ['액션', '모험', '판타지', '무협', 'SF']:
        return '액션/어드벤처/SF'
    elif genre in ['드라마', '멜로/로맨스', '코미디', '블랙코미디']:
        return '드라마/로맨스/코미디'
    elif genre in ['범죄', '스릴러', '미스터리', '느와르', '서스펜스', '공포']:
        return '스릴러/공포/미스터리'
    elif genre in ['다큐멘터리', '공연실황', '뮤지컬', '전쟁', '가족', '서사']:
        return '다큐멘터리/공연실황/기타'
    #else:
        #return '기타'

# 값 변경
df_cleaned['genre'] = df_cleaned['genre'].apply(convert_genre)

print(df_cleaned)
```
