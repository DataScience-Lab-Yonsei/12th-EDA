```python
import React from 'react';
import { ScatterChart, Scatter, XAxis, YAxis, ZAxis, CartesianGrid, Tooltip, Legend, Line } from 'recharts';

const generateData = () => {
  const baseData = Array.from({ length: 20 }, (_, i) => ({
    x: i,
    y: i * 1.5 + Math.random() * 5
  }));

  // Add outliers
  baseData.push({ x: 10, y: 50 });
  baseData.push({ x: 15, y: 60 });

  return baseData;
};

const data = generateData();

// Calculate linear regression
const n = data.length;
const sumX = data.reduce((sum, point) => sum + point.x, 0);
const sumY = data.reduce((sum, point) => sum + point.y, 0);
const sumXY = data.reduce((sum, point) => sum + point.x * point.y, 0);
const sumX2 = data.reduce((sum, point) => sum + point.x * point.x, 0);

const slope = (n * sumXY - sumX * sumY) / (n * sumX2 - sumX * sumX);
const intercept = (sumY - slope * sumX) / n;

const linearRegressionLine = [
  { x: 0, y: intercept },
  { x: 20, y: slope * 20 + intercept }
];


const medianRegressionLine = [
  { x: 0, y: 0 },
  { x: 20, y: 30 }
];

const RegressionComparison = () => (
  <ScatterChart width={600} height={400} margin={{ top: 20, right: 20, bottom: 20, left: 20 }}>
    <CartesianGrid />
    <XAxis type="number" dataKey="x" name="X" />
    <YAxis type="number" dataKey="y" name="Y" />
    <ZAxis range={[64, 144]} />
    <Tooltip cursor={{ strokeDasharray: '3 3' }} />
    <Legend />
    <Scatter name="Data Points" data={data} fill="#8884d8" />
    <Line name="Linear Regression" data={linearRegressionLine} stroke="#ff7300" strokeWidth={2} dot={false} />
    <Line name="Median Regression" data={medianRegressionLine} stroke="#82ca9d" strokeWidth={2} dot={false} />
  </ScatterChart>
);

export default RegressionComparison;
```
