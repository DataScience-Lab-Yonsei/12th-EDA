# Kruskal-Wallis test

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_csv('/content/drive/MyDrive/EDA의료/train.csv')

variables = ['activity_average_met', 'activity_cal_active',
             'activity_cal_total', 'activity_daily_movement', 'activity_high',
             'activity_inactive', 'activity_inactivity_alerts', 'activity_low',
             'activity_medium', 'activity_met_min_high', 'activity_met_min_inactive',
             'activity_met_min_low', 'activity_met_min_medium', 'activity_non_wear',
             'activity_rest', 'activity_score', 'activity_score_meet_daily_targets',
             'activity_score_move_every_hour', 'activity_score_recovery_time',
             'activity_score_stay_active', 'activity_score_training_frequency',
             'activity_score_training_volume', 'activity_steps', 'activity_total',
             'sleep_awake', 'sleep_breath_average', 'sleep_deep', 'sleep_duration',
             'sleep_efficiency', 'sleep_hr_average', 'sleep_hr_lowest',
             'sleep_light', 'sleep_midpoint_at_delta',
             'sleep_midpoint_time', 'sleep_onset_latency', 'sleep_period_id',
             'sleep_rem', 'sleep_restless', 'sleep_rmssd', 'sleep_score',
             'sleep_score_alignment', 'sleep_score_deep', 'sleep_score_disturbances',
             'sleep_score_efficiency', 'sleep_score_latency', 'sleep_score_rem',
             'sleep_score_total', 'sleep_temperature_delta',
             'sleep_temperature_deviation', 'sleep_total']


group_col = 'target'


kruskal_results = []
```

```python
for var in variables:
    groups = [df[df[group_col] == target][var].dropna() for target in df[group_col].unique()]
    h_stat, p_val = stats.kruskal(*groups)
    kruskal_results.append([var, h_stat, p_val])


kruskal_df = pd.DataFrame(kruskal_results, columns=['Variable', 'H-statistic', 'p-value'])

kruskal_df
```

H0
 : 모든 그룹의 중앙값은 서로 같다.

H1
 : 모든 그룹의 중앙값이 전부 같은 것은 아니다.

```python
significant_kruskal_df = kruskal_df[kruskal_df['p-value'] < 0.05]
significant_kruskal_df.reset_index(drop=True, inplace=True)
significant_kruskal_df
```

# Dunn's test 사후분석

```python
pip install scikit-posthocs
```

```python
import pandas as pd
import scikit_posthocs as sp

df = pd.read_csv('/content/drive/MyDrive/EDA의료/train.csv')


# 유의미한 변수
significant_vars = kruskal_df[kruskal_df['p-value'] < 0.05]['Variable']

group_col = 'target'

dunn_results = {}

# 유의미한 변수 Dunn's Test 수행
for var in significant_vars:
    data = []
    labels = []
    for target in df[group_col].unique():
        group_data = df[df[group_col] == target][var].dropna()
        if len(group_data) > 0:
            data.extend(group_data)
            labels.extend([target] * len(group_data))

    dunn_df = pd.DataFrame({var: data, group_col: labels})


    try:
        dunn_result = sp.posthoc_dunn(dunn_df, val_col=var, group_col=group_col)
        dunn_results[var] = dunn_result
    except Exception as e:
        print(f"Error performing Dunn's test for variable '{var}': {e}")


for var, result in dunn_results.items():
    print(f"\nDunn's Test Results for {var}:")
    print(result)

```

# 한국어 ver

```python
import pandas as pd
import scikit_posthocs as sp


dunn_results = {}

significant_vars = kruskal_df[kruskal_df['p-value'] < 0.05]['Variable']


for var in significant_vars:
    df_subset = []
    for target_kr in df[group_col].unique():
        group_data = df[df[group_col] == target_kr][var].dropna()
        df_subset.append(pd.DataFrame({'value': group_data, 'group': target_kr}))


    combined_df = pd.concat(df_subset, ignore_index=True)


    dunn_result = sp.posthoc_dunn(combined_df, val_col='value', group_col='group')
    dunn_results[var] = dunn_result
    dunn_result.index = dunn_result.index.map({'CN': '정상', 'Dem': '경도 치매', 'MCI': '치매'})
    dunn_result.columns = dunn_result.columns.map({'CN': '정상', 'Dem': '경도 치매', 'MCI': '치매'})

    dunn_results[var] = dunn_result



```
