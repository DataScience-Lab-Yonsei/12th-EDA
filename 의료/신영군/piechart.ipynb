```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件
file_path = 'filtered_data.csv'
df = pd.read_csv(file_path)

# # 应用过滤条件
# condition_city = df['시도'] == '전국'
# condition_province = df['시군구'] == '전국'
# condition_sex = (df['성별'] == '전체') | (df['성별'] == '남') | (df['성별'] == '여')
# combined_condition = condition_city & condition_sex & condition_province
# filtered_data = df[combined_condition]

# # 构建年龄过滤条件，去除非 "85세이상" 的其他 "xx세이상" 标签
# filtered_condition_age = (~df['연령별'].str.contains('세이상')) | (df['연령별'] == '85세이상')

# # 应用年龄过滤条件
# age_filtered_data = filtered_data[filtered_condition_age]

# 计算每个性别的发病率及比率
from matplotlib import rcParams

# 设置字体为支持韩语的字体
rcParams['font.family'] = 'Malgun Gothic'

# 其他代码
grouped = df.groupby('성별')['치매환자유병률'].sum()
total = grouped.sum()
ratios = grouped / total

colors = {'남': 'skyblue', '여': 'lightpink'}
labels = [f'{sex}: {ratio*100:.1f}% ' for sex, ratio in ratios.items()]

plt.figure(figsize=(8, 8))
plt.pie(grouped, labels=labels, colors=[colors[sex] for sex in grouped.index], autopct='%1.1f%%')
plt.title('60대 노인 치매환자 남녀 비율')
plt.show()

```
