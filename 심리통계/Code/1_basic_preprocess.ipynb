```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
profile_path = '2018-personality-data.csv' # change the csv file path

user_profile = pd.read_csv(profile_path)
user_profile.columns = user_profile.columns.str.strip().str.replace(' ', '')
```

```python
big_5 = ['openness', 'agreeableness', 'emotional_stability',
       'conscientiousness', 'extraversion']



plt.figure(figsize=(18, 10))

for i, col in enumerate(big_5, 1):
    plt.subplot(2, 3, i)
    sns.histplot(user_profile[col].dropna(), kde=True, bins=20, color='blue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel('Score')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
cols_ = ['userid', 'openness', 'agreeableness', 'emotional_stability',
       'conscientiousness', 'extraversion']
user_profile = user_profile[cols_]
user_profile.head()
```

```python
ratings = pd.read_csv('/content/drive/MyDrive/24-2 EDA Project/2018_ratings.csv') # change the csv file path

ratings
```

```python
df = user_profile.merge(ratings, left_on = 'userid', right_on = 'useri', how = 'left')
df
```

```python
df['userid'], unique_ids = pd.factorize(df['userid'])

df.columns = df.columns.str.strip().str.replace(' ', '')
df = df.drop(['useri','tstamp'], axis=1)
```
